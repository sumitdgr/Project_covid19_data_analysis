```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Data Analysis On Confirmed Covid Cases

```python
confirmed_cases = pd.read_csv('time_series_covid_19_confirmed.csv')
```

```python
confirmed_cases
```

```python
confirmed_cases.describe()
```

```python
total_confirmed = confirmed_cases[['Country/Region' , '5/29/21']].sort_values(by='5/29/21', ascending=False)
```

```python
top_ten_confirmed = total_confirmed.head(10)
```

```python
top_ten_confirmed #top 10 countries with confirmed cases
```

```python
# Bar chart on total confirmed vs country.
top_ten_confirmed.plot(x='Country/Region', y = '5/29/21', label = 'Confirmed Cases',kind = 'bar', color = 'Orange')
plt.title('Confirmed Cases', color = 'Purple', family='Consolas', fontsize=30, weight = 'bold', pad = 20 )
plt.xlabel('Country', color='Purple', family='Consolas', fontsize=16, weight='bold')
plt.ylabel('Total Confirmed', color='Purple', family='Consolas', fontsize=16, weight='bold')
plt.minorticks_on()
plt.grid(b=True, which='major', alpha=0.5, c='grey', linestyle='-')
plt.grid(b=True, which='minor', alpha=0.2, c='grey', linestyle='-')
plt.show()
```

### Comparison between confirmed, recovered and deaths. 

```python
covid = pd.read_csv('covid_19_data.csv')
```

```python
covid.head()
```

```python
covid.describe()
```

```python
covid.columns
```

```python
# pie chart for Worldwide total Confirmed, Recovered and Death Cases.
group_size = [sum(covid['Confirmed']),
              sum(covid['Recovered']),
              sum(covid['Deaths'])]
group_labels = ['Confirmed\n' + str(sum(covid['Confirmed'])),
                'Recovered\n' + str(sum(covid['Recovered'])),
                'Deaths\n' + str(sum(covid['Deaths']))]
custom_colors = ['skyblue','yellowgreen','tomato']
plt.figure(figsize = (8,8))
plt.pie(group_size, labels = group_labels, colors = custom_colors)
central_circle = plt.Circle((0,0), 0.4, color = 'white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size = 12)
plt.title('Worldwide Total Confirmed, Recovered and Death Cases', fontsize = 20)
plt.show()
```
